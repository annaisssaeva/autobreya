```python
nltk.download('stopwords')
from nltk.corpus import stopwords

stops = list(stopwords.words('russian'))
```

Делаем словари с частотностяи слов из хорошего и плохого отзыва и убираем стоп-слова, чтобы они не мешали анализировать тексты.

```python
good_lemmas = {}

lemmatizer = rulemma.Lemmatizer()
lemmatizer.load()

tokenizer = rutokenizer.Tokenizer()
tokenizer.load()

tagger = rupostagger.RuPosTagger()
tagger.load()

tokens = tokenizer.tokenize(good)
tags = tagger.tag(tokens)
lemmas = lemmatizer.lemmatize(tags)
for word, tags, lemma, *_ in lemmas:
    if lemma.isalpha():
        if lemma not in stops:
            if lemma in good_lemmas.keys():
                good_lemmas[lemma] += 1
            else:
                good_lemmas[lemma] = 1

print(good_lemmas)
```

```python
bad_lemmas = {}

tokens = tokenizer.tokenize(bad)
tags = tagger.tag(tokens)
lemmas = lemmatizer.lemmatize(tags)
for word, tags, lemma, *_ in lemmas:
    if lemma.isalpha():
        if lemma not in stops:
            if lemma in bad_lemmas.keys():
                bad_lemmas[lemma] += 1
            else:
                bad_lemmas[lemma] = 1

print(bad_lemmas)
```

Теперь положим в множества слова, повторяющиеся больше 2 раз. Принты я оставила для наглядности.

Уберем из множеств слова, которые есть и в плохих, и в хороших отзывах.

```python
only_bad_lemmas = popular_bad_lemmas.difference(popular_good_lemmas)
only_good_lemmas = popular_good_lemmas.difference(popular_bad_lemmas)

print(only_bad_lemmas)
print(only_good_lemmas)
```

Для оценки работы программы возьмем пока совсем игрушечный пример с количеством слов из хороших отзывов заведомо большим, чем из плохих.

```python
test_text = "Это отличный сериал, который должен посмотреть каждый. Советую!"
```

Функция оценки текста:

```python
def g_or_b(otzyv):
    answ = "unknown"
    good_count = 0
    bad_count = 0
    for word in tokenizer.tokenize(otzyv):
        if word in only_bad_lemmas:
            bad_count += 1
        elif word in only_good_lemmas:
            good_count += 1
    if good_count > bad_count:
        answ = "отзыв положительный"
    elif bad_count > good_count:
        answ = "отзыв отрицательный"
    return(answ)
```

```python
print(g_or_b(test_text))
```

Теперь посчитаем на игрушечном примере accuracy. Допустим, изначально нам дано, что отзыв положительный.

```python
from sklearn.metrics import accuracy_score
```

```python
pred = []
true_answ = []

pred.append(g_or_b(test_text))
true_answ.append("отзыв положительный")

print(accuracy_score(pred, true_answ))
```
